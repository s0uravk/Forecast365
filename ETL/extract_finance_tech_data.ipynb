```python
#import dependencies
!pip install yfinance
!pip install pandas_datareader
!pip install pandas
import pandas as pd
import yfinance as yf
yf.pdr_override()
from pandas_datareader import data as wb
from datetime import date
```

```python
#create dataframe to hold the information of Tickers
info_df = pd.DataFrame({"Ticker" : ["WD", "PFSI", "LDI", "GHI", "COOP", "AFL", "HIG", "PRU", "ALL", "PGR", "MS", "BLK", "GS", "TROW", "BEN", "V", "MA", "AXP", "DFS", "COF",
                                    "QCOM", "TXN", "ADI", "MU", "ARM", "CRM", "SAP", "INTU", "NOW", "ZI", "AAPL", "MSFT", "SONY", "LPL", "SONO", "ACN", "IBM", "FI", "INFY", "CTSH"],
                        "Industry" : "",
                        "Sector" : ""})

info_df["Industry"].iloc[0:5] ="Mortgage"
info_df["Industry"].iloc[5:10] ="Insurance"
info_df["Industry"].iloc[10:15] ="Asset Management"
info_df["Industry"].iloc[15:20] ="Credit Services"
info_df["Industry"].iloc[20:25] ="Mortgage"
info_df["Industry"].iloc[25:30] ="Insurance"
info_df["Industry"].iloc[30:35] ="Asset Management"
info_df["Industry"].iloc[35:40] ="Credit Services"

info_df["Sector"].iloc[0:20] = "Finance"
info_df["Sector"].iloc[20:40] = "Technology"

#Display dataframe
info_df
```

```python
#set end date for today's date for up to date data
end = date.today()
#List to append data temporarily
ls= []
#dataframe to hold all the retrieved data
final_df = {}

#loop through each ticker in info_df
for sym in info_df["Ticker"]:

    #store data retrieved a Ticker from 2018-01-01 to today's date
    stock_data = wb.DataReader({sym}, start='2018-1-1', end= end)
    stock_data["Ticker"] = sym

    #appennding the data in ls list
    ls.append(stock_data)

    #concate that ls to final_df dataframe
    final_df = pd.concat(ls)
#show the dataframe
final_df
```

```python
#merging the dataframes to show sector and industry
finance_tech_df = pd.merge(final_df, info_df, on = "Ticker", how = "left")
finance_tech_df["Date"] = final_df.index
finance_tech_df = finance_tech_df[["Date", "Ticker", "Sector", "Industry", "Open", "High", "Low", "Close", "Volume"]]
#show the result
finance_tech_df
```

```python
#create a csvfile
finance_tech_df.to_csv("Datasets/Finance_Tech_data.csv", index = False)
```
