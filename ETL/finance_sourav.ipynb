```python
#import dependencies
# !pip install yfinance
# !pip install pandas_datareader
import pandas as pd
import yfinance as yf
yf.pdr_override()
from pandas_datareader import data as wb
from datetime import date
```

```python
#set end date for today's date for up to date data
end = date.today()

tickers = ["WD", "PFSI", "LDI", "GHI", "COOP", "AFL", "HIG", "PRU", "ALL", "PGR", "MS", "BLK", "GS", "TROW", "BEN", "V", "MA", "AXP", "DFS", "COF"]

#List to append data temporarily
temp_data= []

#dataframe to hold all the retrieved data
final_df = {}

#loop through each ticker in info_df
for ticker in tickers:

    #store data retrieved a Ticker from 2018-01-01 to today's date
    stock_data = wb.DataReader({ticker}, start='2018-1-1', end= end)
    stock_data["Ticker"] = ticker
    
    #appennding the data in ls list
    temp_data.append(stock_data)
    
    #concate that ls to final_df dataframe
    final_df = pd.concat(temp_data)
#show the dataframe
final_df
```

```python
final_df.reset_index()
final_df.to_csv("Financial_stock_data.csv")
```
